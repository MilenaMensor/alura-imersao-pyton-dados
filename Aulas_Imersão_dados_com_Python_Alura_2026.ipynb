```python
df['contrato'] = df['contrato'].replace({
    'FT': 'Tempo Integral',
    'CT': 'Contrato',
    'PT': 'Tempo Parcial',
    'FL': 'Freelancer'
})

df['contrato'].value_counts()
```

```python
df['tamanho_empresa'] = df['tamanho_empresa'].replace({
    'M': 'Média',
    'L': 'Grande',
    'S': 'Pequena'
})

df['tamanho_empresa'].value_counts()
```

```python
df['remoto'] = df['remoto'].replace({
    0: 'Presencial',
    50: 'Híbrido',
    100: 'Remoto'
})

df['remoto'].value_counts()
```

```python
# Irá mostrar os valores únicos da coluna 'ano' que foi passado, podendo ser utilizado para todas as colunas. O 'nan'
# significa Not A Number, ou seja, tem pessoas na base que não possuem preenchido o dado 'ano'.
df['ano'].unique()
```

```python
# Mostra uma varredura feita na base onde o NaN é presente. Assim devolvendo em formato de uma tabela, quem são.
df[df.isnull().any(axis=1)]
```

```python
import numpy as np

# criação de um datafram de teste para usar de exemplo
df_salario = pd.DataFrame({
    'nome': ['Ana', "Bruno", "Carlos", "Daniele", "Val"],
    'salario': [4000, np.nan, 5000, np.nan, 100000]
})

# calcula a média e substitui os nulos pela média e arredonda os valores
df_salario['salario_media'] = df_salario['salario'].fillna(df_salario['salario'].mean().round(2))

# calcula a mediana e substitui os nulos pela mediana
df_salario['salario_mediana'] = df_salario['salario'].fillna(df_salario['salario'].median())

df_salario
```

```python
# criação de um datafra de teste para usar de exemplo
df_temperaturas = pd.DataFrame ({
    "Dia": ['Segunda', 'Terça', 'Quarta', 'Quinta', 'Sexta'],
    "Temperatura": [30, np.nan, np.nan, 28, 27]
})

# preenche as temperaturas com o valor anterior
df_temperaturas["preenchido_ffill"] = df_temperaturas["Temperatura"].ffill()

df_temperaturas
```

```python
# criação de um dataframe de teste para usar de exemplo
df_temperaturas = pd.DataFrame ({
    "Dia": ['Segunda', 'Terça', 'Quarta', 'Quinta', 'Sexta'],
    "Temperatura": [30, np.nan, np.nan, 28, 27]
})

# preenche as temperaturas com o valor posterior bfill(back fill)
df_temperaturas["preenchido_bfill"] = df_temperaturas["Temperatura"].bfill()

df_temperaturas
```

```python
# criação de um dataframe de teste para usar de exemplo
df_cidades = pd.DataFrame({
    "nome": ["Ana", "Bruno", "Carlos", "Daniele", "Val"],
    "cidade": ["São Paulo", np.nan, "Curitiba", np.nan, "Belém"]
})

# irá preencher os nulos com a informação 'não informado'
df_cidades["cidade_preenchida"] = df_cidades["cidade"].fillna("Não Informado")

df_cidades
```

Exercício= Pegar um cargo e gerar uma visualização de salário médio por país, usando plotly

```python
df_limpo.head()
```

```python
df_limpo['cargo'].value_counts()
```

```python
# Filtrar o DataFrame para o cargo 'Data Analyst'
df_data_analyst = df_limpo[df_limpo['cargo'] == 'Data Scientist']

# Calcular o salário médio por localização da empresa para 'Data Analyst'
salario_medio_por_pais = df_data_analyst.groupby('empresa')['usd'].mean().reset_index().sort_values(by='usd', ascending=False)

# Criar o gráfico de barras interativo usando Plotly
fig = px.bar(salario_medio_por_pais,
             x='empresa',
             y='usd',
             title='Salário Médio para Data Analyst por País da Empresa',
             labels={'empresa': 'País da Empresa', 'usd': 'Salário Médio Anual (USD)'},
             color='empresa',
             color_discrete_sequence=px.colors.qualitative.Set3)

fig.show()
```

```python
df_ds = df_limpo[df_limpo['cargo'] == 'Data Scientist']
media_ds = df_ds.groupby('residencia')['usd'].mean().sort_values(ascending=False).reset_index()

plt.figure(figsize=(10, 5))
sns.barplot(data=media_ds.head(5), x='residencia', y='usd')
plt.title('Top 5 Países com maior média Salarial para Cientistas de Dados')
plt.xlabel('País')
plt.ylabel('Média Salarial (USD)')
plt.show()
```

```python
import pycountry

# Função para converter ISO-2 para ISO-3
def iso2_to_iso3(code):
    try:
        return pycountry.countries.get(alpha_2=code).alpha_3
    except:
        return None

# Criar nova coluna com código ISO-3
df_limpo['residencia_iso3'] = df_limpo['residencia'].apply(iso2_to_iso3)

# Calcular média salarial por país (ISO-3)
df_ds = df_limpo[df_limpo['cargo'] == 'Data Scientist']
media_ds_pais = df_ds.groupby('residencia_iso3')['usd'].mean().reset_index()

# Gerar o mapa
fig = px.choropleth(media_ds_pais,
                    locations='residencia_iso3',
                    color='usd',
                    color_continuous_scale='rdylgn',
                    title='Salário médio de Cientista de Dados por país',
                    labels={'usd': 'Salário médio (USD)', 'residencia_iso3': 'País'})

fig.show()
```
